```python
#now we gonna work on while and for loops
#while loop is used to repeat a block of code as long as the condition is true 
count = 0
while count <= 10:
    print ("count is :", count)
    count += 2
#here in the while loop we are printing the count value and incrementing it by 2 until it reaches 10
#here i am gonna use for loop to print the numbers from 0 to 10
for i in range(1,10):
    print("here is the number :", i)
#now i am gonna use for loop to print the numbers from 0 to 10 with step of 2
for i in range(0, 10, 2):
    print("here is the number with the step of two:", i)
#here inside the range function we have used 3 arguments, the first one is the starting point, the second one is the ending point and the third one is the step
#now i am gonna write no backwardd using for loop
for i in range (10, 0, -1):
    print ("here are the no backward:",i)
#now i am gonna use while loop to print no backward 
count =10
while count > 0:
    print("here are the no backward using while loop:", count)
    count -= 1
#here now i am gonna use for loop to print * from taking input  from the user
stars = int (input("no of starts you want to print:"))
for i in range (stars):
    print("*")
#here we gonna use while loop to print * from taking input from the user
stars = int(input("no of stars you want to print:"))
count = 0
while count < stars:
    print("*")
    count +=1
#next we are gonna jump to break  statement in for loop
for i in range(1, 10):
    if i == 5:
        break
    print("here are the no before break statement:", i)
#break statement is used to exit the loop when the condition is met
#now we are gonna use continue statement in for loop
for i in range(1 ,10):
    if i == 8:
        continue
    print("print no without 8;",i)
#continue statement is used to skip the current iteration of the loop when the condition is met
#we are gonna use break and continue statement in while loop
count = 0
while count < 10:
    if count == 5:
        break
    print("here are the no before break statement:", count)
    count += 1
#here we are using continue statement in while loop
count = 0
while count < 10:
    count += 1
    if count == 8:
        continue
    print("print no without 8:",count)
#in while loop we write exit statement in the end of the loop in break statement and normal while loop
#in while loop for writing continue statement we write it in the middle of the loop
```

```python
import os

# Step 1: Show the current working directory
print("📂 Current directory:", os.getcwd())

# Step 2: Define folder path
folder_path = os.path.join(os.getcwd(), "data")

# Step 3: Check if the folder exists, if not create it
if not os.path.exists(folder_path):
    print("🆕 Folder 'data' not found. Creating it...")
    os.mkdir(folder_path)
else:
    print("✅ Folder 'data' already exists.")

# Step 4: Create a file inside the folder
file_path = os.path.join(folder_path, "chetanya.txt")
print(f"📝 Creating file: {file_path}")

with open(file_path, "w") as file:
    file.write("this is my file inside a folder")

print("📄 File written successfully!")

# Step 5: List all files in the 'data' folder
print("\n📁 Files inside 'data' folder:")
for filename in os.listdir(folder_path):
    print("•", filename)
```

```python
#NOW I AM GONNA WORK ON FUNCTION
# WHAT is function?    ans:-function is the named block of code that perform a specific task. Helps in avoiding repeating code by writing it once and calling it whenevr it needed
#BASIC SYNTAX    
#  def function_name(parameter):
    #block of vode inthis place
#return value
#EXAMPLE:-
def greet(name):
    print("hello",name)
greet("chetanya")    
# Defining a basic function in Python using the 'def' keyword.
# A function is a reusable block of code that performs a specific task.
# This function is named 'greet' and takes one parameter: 'name'.
# When called, it prints a personalized greeting using the provided name.
# Calling the function with the argument "Chetanya"
# This will output: Hello Chetanya



#HERE I AM GONNA LEARN PARAMETER VS ARGUMENT
#parameter is the name used inside the function defination 
#argument is the actual value you pass when you call the function

def greet_person(name):#here name is the parameter
    print("hii",name)
greet_person("chetanya")#here chetanya is the argument


#I AM GONNA LEARN RETURN VS PRINT 
# return gives the value back to the caller so it can be stored or reused
# unlike print , return dosen't display the result directly
#EXAMPLE
def square(n):
    return n*n #this function return backs the square of the no using the return keyword

#here we store the returned value in a variable and print it ourselve
result = square(4)
print("square of the no is:", result)

#compare the print inside the function :
def square_print(n):
    print("square of the no:",n * n)#here print will just display the result

square_print(4)#output will be 16 but you can't store it

print("hello",square_print(4))
print("hii", square(3))


#i am gonna learn default parameter
#you can assing default value to a parameter in a function defination
#if no argument is provided the default will be used 
def greet(name="friend"):
    print("hello",name)
greet()#uses default argument
greet("chetanya")#uses passed argument


#NOW I AM GOONA LEARN VARIABLE SCOPE GLOBAL AND LOCAL
#  a local variable is defined inside a function and only exist there 
# a global variable is defined outside and can be accessed anywhere 
x=10    #global variable
def example_scope():
    x=5#local variable (shadows the global x)
    print ("inside function",x)
example_scope ()#give output :- iside function :5 
print("outside function",x)#give output:- outside fubnction :10


#i am gonna learn variable length argument with arguments
# args allow us to pass the variabble number of argument to a function
#inside the function args are treated as a tuple
def add_numbers(*args):
    print ("argument received:",args)
    total = sum (args)
    print("sum",total)
add_numbers(1,2,3,4)
add_numbers(10,20,30.40)


#gonna learn variable length keyword
#this function accepts any no of keyword argument using **kwargs.
# kwargs stand for keyword argument and collect them in to the dictinory
# each argument passed as key=value will be stored as a key value pair inside that dictionary.
def show_info(**kwargs):
    #we loop through each key value pair in the kwarg dictionary using .item
    for key,value in kwargs.items():
        print(value)
#function call with 3 keyword argument.
#these will be picked into the kwags dictionary.
show_info(name="chetanya",age="19",college="jecrc")


```

```python
#writing a program on what i have known through i.e variable, function , conditionals , loops , exception handling
#NUMBER GUESSING GAME
import random  # We use import random to generate a random number
#outer loop to allow replaying the game
while True:
    num = random.randint(1,5)#generates a secret nowhich get stored in num
    count = 0

    #inner loop keep asking input from user until correct one is guessed
    while True:
        try:
            num1 = int(input("enter the no you guessed"))
            count += 1#increses guess count on each valid input

            if num1 == num:
                print (f"you have guessed it right in {count}count.")
                break#exit the inner loop if the no guessed is correct
            elif num1 < num:
                print("too low")
            else:
                print("too high")

        except ValueError:
            print("please enter a valid value")

    #ask user if wants to play again
    play = input("do you want to replay this game (yes/no)").strip().lower()
    #here strip()removes any leading or trailing spaces from input(ex " yes ")
    #lower() converts input to lowercase so YES Yes can be treated same
    if play == "no":
        print("thank you for playing this game.")
        break
```
